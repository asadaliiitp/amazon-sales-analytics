```python
df.columns = (df.columns.str.lower())
```

```python
df.columns
```

```python
df['orderdate'] = pd.to_datetime(df['orderdate'], errors = 'coerce')
```

```python
num_columns = ['quantity', 'unitprice', 'discount', 'tax', 'shippingcost', 'totalamount']

for col in num_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.dtypes
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.select_dtypes(include='object').columns
```

```python
for col in df.select_dtypes(include='object'):
    df[col] = df[col].str.strip()
```

```python
df['city'].unique()[:10]
```

```python
df['order_year'] = df['orderdate'].dt.year
df['order_month'] = df['orderdate'].dt.month
df['order_quarter'] = df['orderdate'].dt.to_period('Q').astype(str)
```

```python
df['gross_amount'] = df['quantity'] * df['unitprice']
```

```python
df['discount_amount'] = df['gross_amount'] * df['discount']
```

```python
df['net_sales'] = df['gross_amount'] - df['discount_amount']
```

```python
df[['gross_amount', 'discount_amount', 'net_sales']].head()
```

```python
df.to_csv('C:/Users/ASUS/Downloads/Amazon_cleaned.csv', index=False)
```
